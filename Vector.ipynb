```python
import numpy as np

# Define a tolerance for floating-point comparisons
tolerance = 1e-9

while True:
    # Initialize two 3-dimensional vectors with random values between -1 and 1
    vector1 = np.random.uniform(-1,1,3)
    vector2 = np.random.uniform(-1,1,3)

    # Calculate the dot product
    dot_product = np.dot(vector1, vector2)

    # Check for orthogonality (dot product is approximately zero)
    if abs(dot_product) < tolerance:
        print("Found orthogonal vectors!")
        print(f"Vector 1: {vector1}")
        print(f"Vector 2: {vector2}")
        print(f"Dot product: {dot_product:.10f} (approximately zero)")
        break # Exit the loop if vectors are orthogonal
    else:
        print(f"Vectors are not orthogonal. Dot product: {dot_product:.10f}")
        # Optional: Print vectors if not orthogonal to see the values
        # print(f"Vector 1: {vector1}")
        # print(f"Vector 2: {vector2}")

print("Program finished.")
```
