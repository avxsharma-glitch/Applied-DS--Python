```python
import numpy as np

# 1. Create two vectors using the random function
vector1 = np.random.rand(5)  # A 5-element random vector
vector2 = np.random.rand(5)  # Another 5-element random vector

print("Vector 1:", vector1)
print("Vector 2:", vector2)

# 2. Merge them into a single matrix using the column_stack function
matrix = np.column_stack((vector1, vector2))

print("\nMatrix created using column_stack:")
print(matrix)

# 3. Decompose the matrix into Q and R matrices using QR decomposition
Q, R = np.linalg.qr(matrix)

print("\nQ matrix (orthogonal matrix):")
print(Q)

print("\nR matrix (upper triangular matrix):")
print(R)

# Verify decomposition (optional: A = Q @ R)
# print("\nVerification (Q @ R):")
# print(Q @ R)
```
