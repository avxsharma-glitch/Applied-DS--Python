```python
import networkx as nx
import matplotlib.pyplot as plt

# Marvel Character as Vertices
edges = [
    ("Iron man","Spiderman","Mentor"),
    ("Caption America" , "Iron man" , "Teammate"),
    ("Thor" , "Hulk" , "Battle Patner"),
    ("Doctor Strange", "Thor", "Supports"),
    ("Hulk","Caption America","Team Avengrs"),
    ("Spiderman","Doctor Strange",("Allies"))
]
#Create Graphs
G = nx.Graph()

# Add edges with the relationship Labels
for source , target , relationship in edges:
    G.add_edge(source,target,label=relationship)

#Layout
pos = nx.spring_layout(G,seed = 42)

#Draw Graph
plt.figure(figsize=(10,10))
nx.draw(G,pos,with_labels=True, node_size=3000 , font_size=11,font_weight="bold")

# Edge labels
labels = nx.get_edge_attributes( G ,'label')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)

plt.title("Marvel Character Relationship Graph ")
plt.show()
```
